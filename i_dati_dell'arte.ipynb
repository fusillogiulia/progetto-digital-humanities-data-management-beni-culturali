IL CONFRONTO

Creo un grafico di confronto tra i generi più utilizzati dai pittori e dalle pittrici

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/dhdmch/2025-2026/refs/heads/main/data/vapod/data.csv"
df = pd.read_csv(url)
```

Pulisco la colonna dei generi e trasformo in stringa

```python
df['generi_puliti'] = df['generi'].astype(str).str.strip()
if df['generi_puliti'].replace('nan', '').str.len().sum() == 0:
    df['generi_puliti'] = df['contenuti'].astype(str).str.strip()
```

Divido i soggetti in donne e uomini

```python
df_donne = df[df['artisti'].str.contains('femmina', case=False, na=False)]
df_uomini = df[df['artisti'].str.contains('maschio', case=False, na=False)]
```

Escludo nan dai conteggi

```python
conteggio_donne = df_donne['generi_puliti'].value_counts()
if 'nan' in conteggio_donne:
    conteggio_donne = conteggio_donne.drop('nan')
```

```python
conteggio_uomini = df_uomini['generi_puliti'].value_counts()
if 'nan' in conteggio_uomini:
    conteggio_uomini = conteggio_uomini.drop('nan')
```

Prendo in considerazione i 6 generi più utilizzati

```python
dati_donne = conteggio_donne.head(6)
dati_uomini = conteggio_uomini.head(6)
```

Creo un grafico per i pittori e uno per le pittrici

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

# colori inclusivi
colore_donne = '#E66101'
colore_uomini = '#5E3C99'

# grafico donne
ax1.barh(dati_donne.index, dati_donne.values, color=colore_donne, edgecolor='black', height=0.5)
ax1.set_title("Top 6 Generi / Soggetti - PITTRICI", fontsize=16, pad=15, weight='bold', color=colore_donne)
ax1.set_xlabel("Numero di Opere", fontsize=12, weight='bold')
ax1.tick_params(axis='both', labelsize=12)
ax1.invert_yaxis()
ax1.grid(axis='x', linestyle='--', alpha=0.4)
# numeri barre
for i, v in enumerate(dati_donne.values):
    ax1.text(v + 0.1, i, f" {int(v)}", va='center', fontsize=11, weight='bold')

# grafico uomini
ax2.barh(dati_uomini.index, dati_uomini.values, color=colore_uomini, edgecolor='black', height=0.5)
ax2.set_title("Top 6 Generi / Soggetti - PITTORI", fontsize=16, pad=15, weight='bold', color=colore_uomini)
ax2.set_xlabel("Numero di Opere", fontsize=12, weight='bold')
ax2.tick_params(axis='both', labelsize=12)
ax2.invert_yaxis()
ax2.grid(axis='x', linestyle='--', alpha=0.4)

# numeri barre
for i, v in enumerate(dati_uomini.values):
    ax2.text(v + 1, i, f" {int(v)}", va='center', fontsize=11, weight='bold')

# titolo
plt.suptitle("Analisi Comparativa dei Soggetti Artistici per Genere", fontsize=20, weight='bold', y=1.02)

# mostra
plt.tight_layout()
plt.savefig("confronto_affiancato_pulito.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
coordinate_musei = {
    'Pinacoteca Nazionale di Bologna': [44.4979, 11.3524],
    'Galleria degli Uffizi': [43.7677, 11.2556],
    'Palazzo degli Uffizi': [43.7677, 11.2556],
    'Corridoio Vasariano': [43.7677, 11.2556],
    'Museo di San Domenico': [44.2214, 12.0358],
    'Galleria d\'arte moderna di Firenze': [4.4556, 11.1501],
    'Accademia di belle arti di Brera': [44.4979, 11.3524],
    'Palazzo Pitti': [43.7651, 11.2501],
    'sala 2.4 della Pinacoteca Civica di palazzo Volpi': [45.8073, 9.0833]
}
```

Raggruppo le opere per museo, creo il testo pop up e i marker

```python
for museo, gruppo in df_pittrici.groupby('collezioni'):
    gruppo_top_5 = gruppo[gruppo['artisti'].isin(top_5_nomi)]

    if not gruppo_top_5.empty and museo.strip() in coordinate_musei:
        testo_popup = f"<h3><b>{museo}</b></h3><p>Opere delle top 5 pittrici:</p><ul>"

        for index, riga in gruppo_top_5.iterrows():
            artista = riga['artisti']
            titolo = riga.get('titolo', 'Opera d\'arte')
            testo_popup += f"<li><b>{artista}</b> - <i>{titolo}</i></li>"

        testo_popup += "</ul>"

        folium.Marker(
            location=coordinate_musei[museo.strip()],
            popup=folium.Popup(testo_popup, max_width=350),
            tooltip=f"Clicca per vedere le opere delle top 5 pittrici in questo museo",
            icon=folium.Icon(color='blue', icon='star', prefix='fa')
        ).add_to(mappa_top_5)
```
